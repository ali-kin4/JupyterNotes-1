```python
loc = 'D:\Stocks'
os.chdir(loc)
df = pd.read_csv('Code Name Data.csv')
```

```python
a = []
active_codes = []
for i in range(len(df.Codes)):
    try:
        a.append(r.text.index(str(df.Codes[i])))
        active_codes.append(df.Codes[i])
    except ValueError:
        print("Value not found")
```

```python
dd = []
for i in range(len(a)):
    try:
        dd.append(r.text[a[i]+len(str(active_codes[i])):a[i]+80])
    except ValueError:
        print("Value not found")
```

```python
for i in range(len(dd)):
    dd[i] = dd[i].replace("</a></td>\r\n", '')
    dd[i] = dd[i].replace("\u200c", ' ')
    dd[i] = dd[i].replace("title=", ' ')
    dd[i] = dd[i].replace("                  ", '')
    dd[i] = dd[i].replace('"', '')
    dd[i] = dd[i].replace(">", ' ')
    dd[i] = dd[i].replace("<td", '')
    dd[i] = dd[i].replace("<td 44", '')
    dd[i] = dd[i].replace("</a \r\n", '')
    dd[i] = dd[i].replace("</t", '')
    dd[i] = dd[i].replace("</", '')
    dd[i] = dd[i].replace("<", '')
    dd[i] = dd[i].replace("44", '')
    dd[i] = dd[i].replace("(", '')
    dd[i] = dd[i].replace(")", '')
```

```python
names = []
for i in range(len(dd)):
    names.append(dd[i].split()[-1])
    print(str(active_codes[i]) + ": " + dd[i].split()[-1])
```

```python
code_data = {"Names": names,
             "Codes": active_codes}

code_dataF = pd.DataFrame(code_data)

# code_dataF.set_index(code_dataF["Names"], inplace=True)
```
