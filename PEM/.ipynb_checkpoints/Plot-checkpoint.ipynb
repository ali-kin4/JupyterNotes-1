```python
import os
import csv
import pylab as pp
#from scipy.interpolate import spline
import numpy as np
```

```python
os.chdir("D:\PEM article\PEM_Base\Data")

fileName = "V_I_validation.png"
```

```python
x1 = []
y1 = []

x2 = []
y2 = []

x3 = []
y3 = []

x4 = []
y4 = []
```

```python
with open('1phase.txt','r') as csvfile:
    Data1 = csv.reader(csvfile, delimiter=',')
    for row in Data1:
        x1.append(row[0])
        y1.append(row[1])

for i in range(len(x1)):
    x1[i] = float(x1[i])
    y1[i] = float(y1[i])
#----------------------------------------------------------#
with open('2phase.txt','r') as csvfile:
    Data2 = csv.reader(csvfile, delimiter=',')
    for row in Data2:
        x2.append(row[0])
        y2.append(row[1])

for i in range(len(x2)):
    x2[i] = float(x2[i])
    y2[i] = float(y2[i])
    
# x2s = np.linspace(min(x2), max(x2), 1000)
# y2s = spline(x2, y2, x2s)
#----------------------------------------------------------#
with open('Ahmadi.txt','r') as csvfile:
    Data3 = csv.reader(csvfile, delimiter=',')
    for row in Data3:
        x3.append(row[0])
        y3.append(row[1])

for i in range(len(x3)):
    x3[i] = float(x3[i])
    y3[i] = float(y3[i])  
#----------------------------------------------------------#
with open('wang.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x4.append(row[0])
        y4.append(row[1])

for i in range(len(x4)):
    x4[i] = float(x4[i])
    y4[i] = float(y4[i])
```

```python
%matplotlib notebook

pp.autoscale(enable=True, axis='x', tight=True)

pp.plot(x1, y1, marker='o', markerfacecolor='blue', markersize=3, color='skyblue', linewidth=2, label = "Numerical results (single-phase)")
pp.plot(x2, y2, marker='<', markersize = 2.3, markevery=15, color='red', linewidth=1.5, label = "Numerical results (two-phase)")
pp.plot(x3, y3, marker='', color='green', linewidth=2, linestyle='dashed', label = "Ahmadi et al. [24]")
pp.plot(x4, y4, marker='', color='olive', linewidth=1.5, label = "Wang et al. [25]")

#pp.legend("Numerical results (single phase)", "Numerical results (two phase)", "Ahmadi et al.", "Wang et al.")

pp.legend()

pp.xlabel(r"Current density $\left(\frac{A}{cm^2}\right)$")
pp.ylabel(r"Cell voltage $(V)$")


fig = pp.gcf()
fig.set_size_inches(6, 5)
```

```python
pp.savefig(fileName, dpi = 1200)
```
