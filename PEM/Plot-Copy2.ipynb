```python
import osz
import csv
import pylab as pp
import numpy as np
```

```python
os.chdir("D:\PEM article\grid independance")
fileName = "V_I grid2.png"
```

```python
x1 = []
y1 = []

x2 = []
y2 = []

x3 = []
y3 = []

x4 = []
y4 = []

x5 = []
y5 = []
```

```python
with open('50k.txt','r') as csvfile:
    Data1 = csv.reader(csvfile, delimiter=',')
    for row in Data1:
        x1.append(row[0])
        y1.append(row[1])

for i in range(len(x1)):
    x1[i] = float(x1[i])
    y1[i] = float(y1[i])
#----------------------------------------------------------#
with open('100k.txt','r') as csvfile:
    Data2 = csv.reader(csvfile, delimiter=',')
    for row in Data2:
        x2.append(row[0])
        y2.append(row[1])

for i in range(len(x2)):
    x2[i] = float(x2[i])
    y2[i] = float(y2[i])
#----------------------------------------------------------#
with open('200k.txt','r') as csvfile:
    Data3 = csv.reader(csvfile, delimiter=',')
    for row in Data3:
        x3.append(row[0])
        y3.append(row[1])

for i in range(len(x3)):
    x3[i] = float(x3[i])
    y3[i] = float(y3[i])  
#----------------------------------------------------------#
with open('400k.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x4.append(row[0])
        y4.append(row[1])

for i in range(len(x4)):
    x4[i] = float(x4[i])
    y4[i] = float(y4[i])  
```

```python
os.chdir("D:\PEM article\PEM_Base\Data")

with open('wang.txt','r') as csvfile:
    Data5 = csv.reader(csvfile, delimiter=',')
    for row in Data5:
        x5.append(row[0])
        y5.append(row[1])

for i in range(len(x5)):
    x5[i] = float(x5[i])
    y5[i] = float(y5[i])  
```

```python
%matplotlib notebook

pp.autoscale(enable=True, axis='x', tight=True)

pp.plot(x1, y1, marker='o', markerfacecolor='blue', markersize=3, color='skyblue', linewidth=3, label = "50k")
pp.plot(x2, y2, marker='4', color='orange', linewidth=1.5, label = "100k")
pp.plot(x3, y3, marker='', color='red', linewidth=2, label = "200k")
pp.plot(x4, y4, marker='>', color='blue', markersize=3, linewidth=1, label = "400k")
pp.plot(x5, y5, marker='+', color='k', linewidth=1.5, label = "Wang's model [25]")


pp.legend();

pp.xlabel(r"Current density $\left(\frac{A}{cm^2}\right)$");
pp.ylabel(r"Cell voltage $(V)$");

fig = pp.gcf()
fig.set_size_inches(6, 5)
```
