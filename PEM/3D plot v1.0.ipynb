```python
import pylab as pp
import os
import csv
import numpy as np
os.chdir('D:/Engineering/PEMFC/Rhombic/A/0.4')
fname = "A.csv"
```

```python
x = []
y = []
z = []
```

```python
with open(fname, 'r') as file:
    data = csv.reader(file, delimiter=',')
    for row in data:
        x.append(row[1])
        y.append(row[2])
        z.append(row[4])

for i in range(len(x)):
    x[i] = float(x[i])
    y[i] = float(y[i])
    z[i] = float(z[i])
```

```python
P = np.ones((len(y), len(x)))
for n in range(len(x)):
    for i in x:
        for j in y:
            P[j, i] = z[n]
```
