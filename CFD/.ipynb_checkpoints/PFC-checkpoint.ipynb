```python
import os
import csv
import pylab as pp
from scipy.interpolate import spline
import numpy as np
import matplotlib
```

```python
os.chdir("D:\PEM article\V_I")

FileName  ="PFC.png"
```

```python
x1  = []
y1  = []

x2  = []
y2  = []

x3  = []
y3  = []

x4  = []
y4  = []

x5  = []
y5  = []

Aeff = 1e-4
```

```python
with open('A.txt','r') as csvfile:
    Data1 = csv.reader(csvfile, delimiter=',')
    for row in Data1:
        x1.append(row[1])
        y1.append(row[0])

y1s = np.zeros(len(x1))

for i in range(len(x1)):
    x1[i] = float(x1[i])
    y1[i] = float(y1[i])
    y1s[i] = y1[i] * x1[i] #* Aeff
#----------------------------------------------------------#
with open('B.txt','r') as csvfile:
    Data2 = csv.reader(csvfile, delimiter=',')
    for row in Data2:
        x2.append(row[1])
        y2.append(row[0])

y2s = np.zeros(len(x2))

for i in range(len(x2)):
    x2[i] = float(x2[i])
    y2[i] = float(y2[i])
    y2s[i] = y2[i] * x2[i] #* Aeff
#----------------------------------------------------------#
with open('C - Copy.txt','r') as csvfile:
    Data3 = csv.reader(csvfile, delimiter=',')
    for row in Data3:
        x3.append(row[1])
        y3.append(row[0])

y3s = np.zeros(len(x3))
        
for i in range(len(x3)):
    x3[i] = float(x3[i])
    y3[i] = float(y3[i])
    y3s[i] = y3[i] * x3[i] #* Aeff
#----------------------------------------------------------#
with open('Base.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x4.append(row[0])
        y4.append(row[1])

y4s = np.zeros(len(x4))
        
for i in range(len(x4)):
    x4[i] = float(x4[i])
    y4[i] = float(y4[i])
    y4s[i] = y4[i] * x4[i] #* Aeff
```

```python
%matplotlib notebook

pp.autoscale(enable=True, axis='x', tight=True)

pp.plot(y1, y1s, marker='o', markerfacecolor='blue', markersize=4, color='skyblue', linewidth=3, label = "Model A")
pp.plot(y2, y2s, marker='*', color='red', linewidth=1.5, label = "Model B")
pp.plot(y3, y3s, marker='', color='green', linewidth=2, linestyle='dashed', label = "Model C")
pp.plot(y4, y4s, marker='<', color='olive', linewidth=1.5, label = "Base model", markevery = 6)
#pp.plot(x5, y5, marker='*', color='blue', linewidth=1.5, label = "Base model")

pp.legend();

pp.xlabel(r"Cell Voltage $(V)$");
pp.ylabel(r"Produced Power Density [$\times 10^{-4}$] $\left(\frac{W}{cm^2}\right)$");


fig = matplotlib.pyplot.gcf()
fig.set_size_inches(6, 5)
```

```python
pp.savefig(FileName, dpi=600)
```
