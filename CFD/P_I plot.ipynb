```python
import os
import csv
import pylab as pp
from scipy.interpolate import spline
import numpy as np
import matplotlib
```

```python
os.chdir("D:\PEM article\V_I")
```

```python
x1 = []
y1 = []

x2 = []
y2 = []

x3 = []
y3 = []

x4 = []
y4 = []

x5 = []
y5 = []
```

```python
with open('A.txt','r') as csvfile:
    Data1 = csv.reader(csvfile, delimiter=',')
    for row in Data1:
        x1.append(row[1])
        y1.append(row[0])

for i in range(len(x1)):
    x1[i] = float(x1[i])
    y1[i] = float(y1[i]) * x1[i]
#----------------------------------------------------------#
with open('B.txt','r') as csvfile:
    Data2 = csv.reader(csvfile, delimiter=',')
    for row in Data2:
        x2.append(row[1])
        y2.append(row[0])

for i in range(len(x2)):
    x2[i] = float(x2[i])
    y2[i] = float(y2[i]) * x2[i]
#----------------------------------------------------------#
with open('C - Copy.txt','r') as csvfile:
    Data3 = csv.reader(csvfile, delimiter=',')
    for row in Data3:
        x3.append(row[1])
        y3.append(row[0])

for i in range(len(x3)):
    x3[i] = float(x3[i])
    y3[i] = float(y3[i]) * x3[i]
#----------------------------------------------------------#
with open('Base.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x4.append(row[0])
        y4.append(row[1])

for i in range(len(x4)):
    x4[i] = float(x4[i])
    y4[i] = float(y4[i]) * x4[i]
```

```python
os.chdir("D:\PEM article\P_I")

with open('Base P_I.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x5.append(row[0])
        y5.append(row[1])
        
for i in range(len(x5)):
    x5[i] = float(x5[i])
    y5[i] = float(y5[i]) * x5[i]
```

```python
%matplotlib notebook

pp.autoscale(enable=True, axis='x', tight=True)

pp.plot(x1, y1, marker='o', markerfacecolor='blue', markersize=4, color='skyblue', linewidth=3, label = "Model A")
pp.plot(x2, y2, marker='*', color='red', linewidth=1.5, label = "Model B")
pp.plot(x3, y3, marker='', color='green', linewidth=2, linestyle='dashed', label = "Model C")
pp.plot(x4, y4, marker='', color='olive', linewidth=1.5, label = "Base model")
#pp.plot(x5, y5, marker='*', color='blue', linewidth=1.5, label = "Base model")

pp.legend();

pp.xlabel(r"Current density $\left(\frac{A}{cm^2}\right)$");
pp.ylabel(r"Power Density $\left(\frac{W}{cm^2}\right)$");


fig = matplotlib.pyplot.gcf()
fig.set_size_inches(6, 5)
```
