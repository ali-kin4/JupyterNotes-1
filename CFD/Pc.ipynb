```python
import os
import csv
import pylab as pp
import numpy as np
import matplotlib.patches as mp
```

```python
os.chdir("D:\PEM article\DeltaP")
fileName = "Pc.png"
```

```python
x1 = []
y1 = []

x2 = []
y2 = []

x3 = []
y3 = []

x4 = []
y4 = []

#------------------------------

u_base = 1.18E-01
uA = 2.73E-02
uB = 6.49E-02
uC = 2.67E-01

#------------------------------

A_base = 1e-6
AA = 4.31e-6
AB = 1.81e-6
AC = 0.44e-6
```

```python
with open('A.txt','r') as csvfile:
    Data1 = csv.reader(csvfile, delimiter=',')
    for row in Data1:
        x1.append(row[0])
        y1.append(row[1])

PcA = np.zeros(len(x1))

for i in range(len(x1)):
    x1[i] = float(x1[i])
    y1[i] = float(y1[i])
    PcA[i] = y1[i] * uA * AA
#----------------------------------------------------------#
with open('B.txt','r') as csvfile:
    Data2 = csv.reader(csvfile, delimiter=',')
    for row in Data2:
        x2.append(row[0])
        y2.append(row[1])

PcB = np.zeros(len(x2))
        
for i in range(len(x2)):
    x2[i] = float(x2[i])
    y2[i] = float(y2[i])
    PcB[i] = y2[i] * uB * AB
#----------------------------------------------------------#
with open('C.txt','r') as csvfile:
    Data3 = csv.reader(csvfile, delimiter=',')
    for row in Data3:
        x3.append(row[0])
        y3.append(row[1])

PcC = np.zeros(len(x3))
        
for i in range(len(x3)):
    x3[i] = float(x3[i])
    y3[i] = float(y3[i])
    PcC[i] = y3[i] * uC * AC
#----------------------------------------------------------#
with open('Base.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x4.append(row[0])
        y4.append(row[1])

Pc_base = np.zeros(len(x4))
        
for i in range(len(x4)):
    x4[i] = float(x4[i])
    y4[i] = float(y4[i])
    Pc_base[i] = y4[i] * u_base * A_base
```

```python
%matplotlib notebook

pp.autoscale(enable=True, axis='x', tight=True)

N = 4
x = np.arange(N) 
width = 0.15
#A
Y1 = [PcA[0], PcA[1], PcA[2], PcA[3]]

#B
Y2 = [PcB[0], PcB[1], PcB[2], PcB[3]]

#C
Y3 = [PcC[0], PcC[1], PcC[2], PcC[3]]

#base
Y4 = [Pc_base[0], Pc_base[9], Pc_base[18], Pc_base[27]]

pp.bar(x, Y1, width, color = 'green', zorder = 2)
pp.bar(x + width, Y2, width, color = 'red', zorder = 2)
pp.bar(x + width*2, Y3, width, color = 'orange', zorder = 2)
pp.bar(x + width*3, Y4, width, color = 'blue', zorder = 2)

pp.xticks(x + width*1.5, ["0.4V", "0.5V", "0.6V", "0.7V"])
pp.xlabel("Cell Voltage $(V)$")
pp.ylabel(r"Consumed Power Density$\left(\frac{W}{cm^2}\right)$");

g1 = mp.Patch(color = 'green', label = "Model A")
g2 = mp.Patch(color = 'red', label = "Model B")
g3 = mp.Patch(color = 'orange', label = "Model C")
g4 = mp.Patch(color = 'blue', label = "Base Model")

pp.legend(handles = [g1, g2, g3, g4], loc='best')

fig = pp.gcf()
fig.set_size_inches(6, 5)
```
