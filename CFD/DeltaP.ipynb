```python
import os
import csv
import pylab as pp
import numpy as np
```

```python
os.chdir("D:\PEM article\DeltaP")
fileName = "DeltaP.png"
```

```python
x1 = []
y1 = []

x2 = []
y2 = []

x3 = []
y3 = []

x4 = []
y4 = []
```

```python
with open('A.txt','r') as csvfile:
    Data1 = csv.reader(csvfile, delimiter=',')
    for row in Data1:
        x1.append(row[0])
        y1.append(row[1])

for i in range(len(x1)):
    x1[i] = float(x1[i])
    y1[i] = float(y1[i])
#----------------------------------------------------------#
with open('B.txt','r') as csvfile:
    Data2 = csv.reader(csvfile, delimiter=',')
    for row in Data2:
        x2.append(row[0])
        y2.append(row[1])

for i in range(len(x2)):
    x2[i] = float(x2[i])
    y2[i] = float(y2[i])
#----------------------------------------------------------#
with open('C.txt','r') as csvfile:
    Data3 = csv.reader(csvfile, delimiter=',')
    for row in Data3:
        x3.append(row[0])
        y3.append(row[1])

for i in range(len(x3)):
    x3[i] = float(x3[i])
    y3[i] = float(y3[i])  
#----------------------------------------------------------#
with open('Base.txt','r') as csvfile:
    Data4 = csv.reader(csvfile, delimiter=',')
    for row in Data4:
        x4.append(row[0])
        y4.append(row[1])

for i in range(len(x4)):
    x4[i] = float(x4[i])
    y4[i] = float(y4[i])  
```

```python
%matplotlib notebook

pp.autoscale(enable=True, axis='x', tight=True)

pp.plot(x1, y1, marker='o', markerfacecolor='blue', markersize=3, color='skyblue', linewidth=3, label = "Model A")
pp.plot(x2, y2, marker='4', color='red', linewidth=1.5, label = "Model B")
pp.plot(x3, y3, marker='', color='green', linewidth=2, label = "Model C")
pp.plot(x4, y4, marker='>', color='olive', markersize=3, linewidth=1, label = "Base Model")


pp.legend();

pp.xlabel(r"Cell Voltage $(V)$");
pp.ylabel(r"Pressure Difference $(Pa)$");

fig = pp.gcf()
fig.set_size_inches(6, 5)
```

```python
pp.savefig(fileName, dpi=1200)
```
