```python
df1 = pd.read_csv("90_new.csv")
df2 = pd.read_csv("180_new.csv")
df3 = pd.read_csv("270_main.csv")
df4 = pd.read_csv("base_new.csv")
```

```python
os.chdir("D:/Engineering/PEMFC/SERP/Data/Result/PC/v")
```

```python
ddf1 = pd.read_csv("90R_new.csv")
ddf2 = pd.read_csv("180R_new.csv")
ddf3 = pd.read_csv("270R_new.csv")
ddf4 = pd.read_csv("base_new.csv")
```

```python
A1 = 1.10E-03

v1 = ddf1.v.tolist()
v2 = ddf2.v.tolist()
v3 = ddf3.v.tolist()
v4 = ddf4.v.tolist()

Pf1 = ddf1.i * ddf1.v * A1
Pf2 = ddf2.i * ddf2.v * A1
Pf3 = ddf3.i * ddf3.v * A1
Pf4 = ddf4.i * ddf4.v * A1
```

```python
A2 = 8.00E-07
u = 5.55E+00

Pc1 = df1.delP * A2 * u
Pc2 = df2.delP * A2 * u
Pc3 = df3.delP * A2 * u
Pc4 = df4.delP * A2 * u
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

ax1.plot(df1.i, Pc1)
ax1.plot(df2.i, Pc2)
ax1.plot(df3.i, Pc3)
ax1.plot(df4.i, Pc4)
ax1.autoscale(enable=True, axis='x', tight=True)

ax2.plot(ddf1.i, Pf1)
ax2.plot(ddf2.i, Pf2)
ax2.plot(ddf3.i, Pf3)
ax2.plot(ddf4.i, Pf4)
ax2.autoscale(enable=True, axis='x', tight=True)
```

```python
N = 4
x = np.arange(N) 
width = 0.15

#A
Y1 = [Pc1[len(Pc1)/4], 2*Pc1[len(Pc1)/4], 3*Pc1[len(Pc1)/4], 4*Pc1[len(Pc1)/4]]

#B
Y2 = [Pc2[len(Pc2)/4], 2*Pc2[len(Pc2)/4], 3*Pc2[len(Pc2)/4], 4*Pc2[len(Pc2)/4]]

#C
Y3 = [Pc3[len(Pc3)/4], 2*Pc3[len(Pc3)/4], 3*Pc3[len(Pc3)/4], 4*Pc3[len(Pc3)/4]]

#base
Y4 = [Pc4[len(Pc4)/4], 2*Pc4[len(Pc4)/4], 3*Pc4[len(Pc4)/4], 4*Pc4[len(Pc4)/4]]

plt.figure(figsize=(8, 8))
sns.set(font_scale=1.5)
plt.bar(x, Y1, width, color = 'green', zorder = 2)
plt.bar(x + width, Y2, width, color = 'red', zorder = 2)
plt.bar(x + width*2, Y3, width, color = 'orange', zorder = 2)
plt.bar(x + width*3, Y4, width, color = 'blue', zorder = 2)

plt.xticks(x + width*1.5, ["0.4V", "0.5V", "0.6V", "0.7V"])
plt.xlabel("Cell Voltage $(V)$")
plt.ylabel(r"Consumed Power Density$\left(\frac{W}{cm^2}\right)$");

g1 = patch.Patch(color = 'green', label = "Model A")
g2 = patch.Patch(color = 'red', label = "Model B")
g3 = patch.Patch(color = 'orange', label = "Model C")
g4 = patch.Patch(color = 'blue', label = "Base Model")

plt.autoscale(enable=True, axis='x', tight=True)

plt.legend(handles = [g1, g2, g3, g4], loc='best')
plt.title('Power Density Consumption')

fig = plt.gcf()
# fig.set_size_inches(7, 6)

# file = 'D:\Engineering\PEMFC\SERP\Data\Result\PC.png'
plt.savefig('D:\PC.pdf', bbox_inches='tight')
# os.startfile(file)
```
